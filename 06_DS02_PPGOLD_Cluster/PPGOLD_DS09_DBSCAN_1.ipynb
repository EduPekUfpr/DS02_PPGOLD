# Density-Based Clustering

## Objetivo

*   Aprender a usar o algoritmo do DBSCAN do Scikit-learn
---

```python
import numpy as np 
import pandas as pd
from sklearn.cluster import DBSCAN 
from sklearn.datasets import make_blobs 
from sklearn.preprocessing import StandardScaler 
import matplotlib.pyplot as plt 
%matplotlib inline

np.random.seed(0)
```

## Dados

```python
X = pd.read_csv("https://raw.githubusercontent.com/EduPekUfpr/DS02_PPGOLD/main/Data/cluster_data.csv", sep="\t")
plt.scatter(X["x"], X["y"], marker='.')
```

```python
X = StandardScaler().fit_transform(X)
plt.scatter(X[:, 0], X[:, 1], marker='.')
```

## DBSCAN
NÃ³s Usaremos 2 parÃ¢metros para o DBSCAN:

* Epsilon - Medida do Raio
* minimumSamples - quantos Vizinhos por raio



```python
epsilon = 0.15
minimumSamples = 6

db = DBSCAN(eps=epsilon, min_samples=minimumSamples).fit(X)

labels = db.labels_
df = pd.DataFrame(labels)
UniqueLabels = df[0].unique()

UniqueLabels
```

```python
plt.figure(figsize=(18, 12))
for i in UniqueLabels:
    if i != -1:
        plt.scatter(X[labels == i , 0] , X[labels == i , 1] , label = i)
    else: # Print OUTLIERS
        plt.scatter(X[labels == i , 0] , X[labels == i , 1] , label = i, marker="x", s=300, color = 'k')
        
plt.legend()
plt.show()
```
